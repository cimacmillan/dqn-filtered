```python
def resample(baseline_reward, reward_list, frame_list, length):
#     print(reward_list)
#     print(frame_list)
    
    final_reward = []
    current_frame_index = 0
    current_exp = 0
    
    
    for i in range(length):
        exp_index = (frame_list[i] / (length * 1000))
        my_index = (i + 1) / length
        
        index_multiplier = my_index / exp_index
        
        index = i * index_multiplier
        
        if index >= length:
            final_reward.append(baseline_reward[i])
        else:
            final_reward.append(reward_list[i])
        

#     print(len(frame_list))
#     print(frame_list[-1])
    return final_reward

def measure_success(baseline, experiments):

    
    baseline_rewards = np.array([i for i in baseline.repeat_rewards])
    baseline_means = baseline_rewards.mean(axis = 0)
    
    
    overall_score = 0
    for i in experiments:
        
        experiment_rewards = np.array([x for x in i.repeat_rewards])
        
        experiment_means = resample(baseline_means, experiment_rewards.mean(axis = 0),  i.accepted_frames_at_tick, len(i.accepted_frames_at_tick))
        diff = (experiment_means - baseline_means)
        sum_diff = (diff.sum() / len(i.repeat_rewards)) / len(i.repeat_rewards[0])
        overall_score += sum_diff
        
    return overall_score

```

```python
def plot_action_time_graph(labels, experiment_actions):


    for exp_id in range(len(experiment_actions)):

        exp = experiment_actions[exp_id]
        label = labels[exp_id]

        resolution = 50

        fig, ax = plt.subplots(figsize=(5, 3))

        ax.set_title('Action-Time for ' + label)
        ax.legend(loc='upper left')
        ax.set_xlabel('Timestep')
        ax.set_ylabel('Action Distribution')
        fig.tight_layout

        minimum_action = 10000;
        maximum_action = 0;
    

        for experiment in experiment_actions:
            for repeat in experiment:

                if len(repeat) == 0:
                    return;

                minimum = np.min(repeat);
                maximum = np.max(repeat);
                if minimum < minimum_action:
                    minimum_action = minimum
                if maximum > maximum_action:
                    maximum_action = maximum

        experiment = exp[0]
        
        print(len(experiment))

    #     experiment = experiment_actions[0][0]
        x_steps = int(len(experiment) / resolution)
        x_ticks = [(x / x_steps) * 4000 for x in range(x_steps)]

        experiment_fill_list = []

        for x in range(x_steps):
            action_sub_list = experiment[x : (x + resolution)]

            occurences = [percentage_of_occurence(i, action_sub_list) for i in range(minimum_action, maximum_action + 1)]
            occurences_fill = []
            occurence_track = 0;
            for occurence in occurences:
                occurences_fill.append((occurence_track, occurence_track + occurence))
                occurence_track += occurence

            experiment_fill_list.append(occurences_fill);

        fill_between = []
        
        texts = ["noop", "fire", "right", "left", "right+fire", "left+fire"];

        for i in range(minimum_action, maximum_action + 1):
            top_y = []
            bottom_y = []
            for x in range(x_steps):
                top_y.append(experiment_fill_list[x][i][0])
                bottom_y.append(experiment_fill_list[x][i][1])

            ax.fill_between(x_ticks, top_y, bottom_y)
            
            ax.text(1.02, (top_y[-1] + bottom_y[-1]) / 2, texts[i], va='center', ha="left",
        transform=ax.get_yaxis_transform())

```

```python
prog = re.compile("\D*tensor\((?P<action>\d+)\)*\D*")
def readActionArray(file):
    text_file = open(file, "r")
    lines = text_file.readlines()
    text_file.close()
    actions = [];

    for line in lines:
        match = prog.match(line);
        if match is not None:
            actions.append(int(match.group("action")))

    return actions;


def plot_action_histogram(labels, action_lists):
    
    bin_width = 0.8
    bins = range(4)
    
    fig, ax = plt.subplots(figsize=(5, 3))
    for action_list, label in zip(action_lists, labels):
        plot_colour = next(ax._get_lines.prop_cycler)['color']
        ax.hist(action_list, bins=3, color = plot_colour,
                 label=label, alpha=.5, rwidth = bin_width)

    ax.set_title('Action Histogram')
    ax.legend(loc='lower right')
    ax.set_xlabel('Action Type')
    ax.set_ylabel('Probability')
    

#     plt.xticks(np.arange(bin_width/2, 1, bin_width), ("Defensive", "Aggressive", "Both"))
#     plt.xlim(bins[0], bins[-1])
    
    locs, labels = plt.yticks()   
    new_labels = [(int((x / len(action_lists[0])) * 100))/100 for x in locs]
    ax.set_yticklabels(new_labels)
    
    fig.tight_layout()
    
def convert(action_list):
    result = ["Defensive", "Aggressive", "Both"]
    for action in action_list:
        if action == 0 or action == 2 or action == 3:
            result.append("Defensive") 
#             Noop
        
        if action == 1:
            result.append("Aggressive")
#             Fire
            
        if action == 4 or action == 5:
#             result.append("Aggressive")
            result.append("Both") 
#             Move
            
    return result
    
small_filtered_actions = convert(readActionArray("../sim/action6_0.txt") + readActionArray("../sim/action6_1.txt") + readActionArray("../sim/action6_2.txt") + readActionArray("../sim/action6_3.txt") +readActionArray("../sim/action6_4.txt"))
large_filtered_actions = convert(readActionArray("../sim/action7_0.txt") + readActionArray("../sim/action7_1.txt") + readActionArray("../sim/action7_2.txt") + readActionArray("../sim/action7_3.txt") + readActionArray("../sim/action7_4.txt"))
small_unfiltered_actions = convert(readActionArray("../sim/action8_0.txt") + readActionArray("../sim/action8_1.txt") + readActionArray("../sim/action8_2.txt") + readActionArray("../sim/action8_3.txt") + readActionArray("../sim/action8_4.txt"))
large_unfiltered_actions = convert(readActionArray("../sim/action9_0.txt") + readActionArray("../sim/action9_1.txt") + readActionArray("../sim/action9_2.txt") + readActionArray("../sim/action9_3.txt") + readActionArray("../sim/action9_4.txt"))

plot_action_histogram(("Unfiltered", "Filtered"), (small_unfiltered_actions, small_filtered_actions))
plot_action_histogram(("Unfiltered", "Filtered"), (large_unfiltered_actions, large_filtered_actions))
```
